# Text8 Dataset Processing
This notebook downloads the text8 dataset, computes basic statistics, builds a character-level tokenizer, and prepares the data for model training.

```python
import os
import urllib.request
import zipfile
from collections import Counter
import numpy as np
import pandas as pd
from IPython.display import display

# --- 1. Download and Extract ---
url = "http://mattmahoney.net/dc/text8.zip"
data_zip = "text8.zip"
data_txt = "text8"

if not os.path.exists(data_txt):
    print("Downloading text8 dataset...")
    urllib.request.urlretrieve(url, data_zip)
    print("Extracting...")
    with zipfile.ZipFile(data_zip, 'r') as zip_ref:
        zip_ref.extractall()
else:
    print("text8 dataset already exists locally.")
```

## 2. Load Text & Compute Basic Stats
Load the characters, count frequencies, and build basic vocabulary statistics.

```python
# --- 2. Load Text ---
with open(data_txt, 'r') as f:
    text = f.read()

# --- 3. Compute Basic Stats ---
total_chars = len(text)
chars = sorted(list(set(text)))
vocab_size = len(chars)

# Count character frequencies
char_counts = Counter(text)
top_chars = char_counts.most_common(5)

# --- 4. Build Tokenizer (Char-level) ---
stoi = {ch: i for i, ch in enumerate(chars)}
itos = {i: ch for i, ch in enumerate(chars)}

def encode(s):
    return [stoi[c] for c in s]

def decode(l):
    return ''.join([itos[i] for i in l])

# --- 5. Split ---
n = len(text)
train_data = text[:int(n*0.9)]
val_data = text[int(n*0.9):]
```

## 3. View Statistics & Examples
Below summarizes the dataset statistics, the top vocabulary elements, an encoded snippet example, and train/val split sizes.

```python
# Basic Dataset Stats
df_stats = pd.DataFrame([
    {"Metric": "Total Characters", "Value": f"{total_chars:,}"},
    {"Metric": "Vocabulary Size", "Value": vocab_size},
    {"Metric": "Vocabulary", "Value": repr(''.join(chars))},
    {'Metric': "Tokenized Vocabulary", "Value": str(encode(''.join(chars)))},
    {"Metric": "Train Split Size", "Value": f"{len(train_data):,}"},
    {"Metric": "Val Split Size", "Value": f"{len(val_data):,}"},
])

# Top 5 Characters
df_top_chars = pd.DataFrame(top_chars, columns=["Character", "Count"]).replace(" ", "<space>")

# Examples Snippet
sample_length = 51
sample_text = text[:sample_length]
sample_tokens = encode(sample_text)

df_example = pd.DataFrame([
    {"Type": "Raw Text", "Content": f"'{sample_text}'"},
    {"Type": "Tokenized", "Content": str(sample_tokens)},
    {"Type": "Decoded", "Content": f"'{decode(sample_tokens)}'"}
])

display(df_stats.style.set_caption("Text8 Dataset Statistics").hide(axis='index'))
print("\n")
display(df_top_chars.style.set_caption("Top 5 Characters").hide(axis='index'))
print("\n")
display(df_example.style.set_caption("Tokenizer Example").hide(axis='index'))
```
